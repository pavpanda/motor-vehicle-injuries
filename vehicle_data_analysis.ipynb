```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("vehicle_data.xlsx")
df = df[2:108]
df
```

```python
def plot_from_year_range(df, start_year, end_year, column_name, title, xlabel, ylabel, filename):
    plt.style.use('ggplot')
    x = np.array(range(start_year, end_year + 1))
    y = np.array(df[column_name][start_year-1913:end_year-1911+1])
    print(y)
    plt.plot(x, y)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.savefig(filename, bbox_inches='tight')
```

```python
plot_from_year_range(df, 1913, 2018, "Number of deaths", "Number of deaths vs Time", "Time (year)", "Number of deaths", "deaths.png")
```

```python
plot_from_year_range(df, 1913, 2018, " Estimated number of Vehicles (millions)", "Estimated number of vehicles vs Time", "Time (year)", "Estimated number of vehicles (millions)", "numvehicles.png")
```

```python
plot_from_year_range(df, 1923, 2018, "Estimated number of Vehicle miles (billions)", "Estimated number of vehicle miles vs Time", "Time (year)", "Estimated number of vehicle miles (billions)", "vehiclemiles.png")
```

```python
plot_from_year_range(df, 1923, 2018, "Estimated number of Drivers (millions)", "Estimated number of drivers vs Time", "Time (year)", "Estimated number of drivers (millions)", "drivers.png")
```
